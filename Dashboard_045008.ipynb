## **Objective**
The objective for the report based on the recruitment dashboard should be to analyze the recruitment process and candidate status, identify trends in candidate applications, evaluate the effectiveness of different recruitment stages and sources, and provide insights into the reasons for candidate declines and withdrawals. Additionally, the report should aim to assess the success rates of different companies and locations in the recruitment process.

The primary goals include:

**Process Analysis:**

- Evaluate the efficiency of each stage in the recruitment process.
Identify bottlenecks and areas for improvement in the workflow.
Streamline the process for a more effective and timely recruitment cycle.

**Trend Identification:**

- Analyze trends in candidate applications over time.
- Identify patterns in the volume and quality of applications.
- Recognize seasonal or cyclical variations in recruitment activities.

**Source Effectiveness:**

- Evaluate the performance of different recruitment sources (e.g., job boards, referrals, social media).
- Determine which sources yield the highest-quality candidates.
- Optimize resource allocation based on the most effective channels.

**Candidate Status Insights:**

- Investigate reasons for candidate declines and withdrawals.
- Provide insights into common trends or issues leading to candidate disengagement.
- Implement strategies to enhance candidate experience and retention.


## **Data Overview:**

- The dataset contains a total of 182 entries and 22 columns.
The provided dataset encompasses comprehensive information on candidate applications and recruitment stages.
It contains information about candidates, including their application details, recruitment stages, and status. Here are some initial observations from the dataset:

- The dataset includes columns such as Comments, Fullname, Email, Phone_Number, Address, DoB (Date of Birth), Gender, Language, Source, Application_Date, Status, Recruitment_Stages, Company, Location, Decline_Reasons, Phone_Screen_Date, Harver_Test_Date, Interview_Date, Offer_Date, Hiring_Date, Payment_Date, and Modification_Date.

- The "Status" column indicates the current status of the candidates, such as "On Hold", "Whitdraw", and "Rejected".

- The "Decline_Reasons" column provides reasons for candidate decline, such as "Language level insufficient" and "Aceitou outra proposta" (Accepted another offer).
